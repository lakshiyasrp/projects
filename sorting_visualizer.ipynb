```
import random
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def bubble_sort(data, draw_data, speed):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                draw_data(data, ['green' if x == j or x == j + 1 else 'red' for x in range(len(data))])
                time.sleep(speed)
    draw_data(data, ['green' for x in range(len(data))])


def insertion_sort(data, draw_data, speed):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and key < data[j]:
            data[j + 1] = data[j]
            j -= 1
            draw_data(data, ['green' if x == j + 1 or x == i else 'red' for x in range(len(data))])
            time.sleep(speed)
        data[j + 1] = key
        draw_data(data, ['green' if x == j + 1 or x == i else 'red' for x in range(len(data))])
        time.sleep(speed)
    draw_data(data, ['green' for x in range(len(data))])


def selection_sort(data, draw_data, speed):
    for i in range(len(data)):
        min_idx = i
        for j in range(i + 1, len(data)):
            if data[min_idx] > data[j]:
                min_idx = j
            draw_data(data, ['green' if x == min_idx or x == j else 'red' for x in range(len(data))])
            time.sleep(speed)
        data[i], data[min_idx] = data[min_idx], data[i]
        draw_data(data, ['green' if x == i or x == min_idx else 'red' for x in range(len(data))])
        time.sleep(speed)
    draw_data(data, ['green' for x in range(len(data))])



def draw_data(data, colors):
    plt.clf()
    plt.bar(range(len(data)), data, color=colors)
    plt.title("Sorting Visualizer")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.draw()
    plt.pause(0.01)


def main():
    n = 20  # Number of elements to sort
    data = [random.randint(1, 100) for _ in range(n)]  # Generate random data
    speed = 0.1  # Adjust the speed of the visualization

    print("Choose a sorting algorithm:")
    print("1. Bubble Sort")
    print("2. Insertion Sort")
    print("3. Selection Sort")

    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        bubble_sort(data.copy(), draw_data, speed)
    elif choice == '2':
        insertion_sort(data.copy(), draw_data, speed)
    elif choice == '3':
        selection_sort(data.copy(), draw_data, speed)
    else:
        print("Invalid choice.")

    plt.show()

if __name__ == '__main__':
    main()
```
